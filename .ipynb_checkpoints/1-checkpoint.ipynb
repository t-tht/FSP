```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline

update_print = True
```

```python
raw = pd.read_csv("priceData.csv")

date = raw["date"]
px = raw["SPX Index"]

logpx = np.log(px)
```

# 1.1.1

```python
plt.plot(px)
plt.grid()
plt.title("1.1.1 - time series")
if update_print:
    plt.savefig("111_time_series.eps")
```

```python
plt.figure()

plt.plot(logpx)
plt.grid()
plt.title("1.1.1 - log time series")
if update_print:
    plt.savefig("111_log_time_series.eps")
```

# 1.1.2

```python
plt.figure()
logpx.rolling(252,min_periods=1).mean().plot()
plt.grid()
plt.title("1.1.2 - rolling mean against time")
if update_print:
    plt.savefig("112_rolling_mean.eps")


plt.figure()
logpx.rolling(252,min_periods=1).std().plot()
plt.grid()
plt.title("1.1.2 - rolling variance against time")
if update_print:
    plt.savefig("112_rolling_var.eps")
```

# 3

```python
logret = logpx.diff()
plt.plot(logret)
plt.grid()
plt.title("1.1.3 - log return")
if update_print:
    plt.savefig("113_logret.eps")

plt.figure()
simpret = px.pct_change()
plt.plot(simpret)
plt.grid()
plt.title("1.1.3 - simple return")
if update_print:
    plt.savefig("113_simpret.eps")
```
