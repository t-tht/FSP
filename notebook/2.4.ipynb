```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## 1. Estimate the market returns per day

```python
df = pd.read_csv(r'../data/fsp_case_31_BSD.csv', index_col =0, header = [0,1])

ret_tot = []

for i in range(157):
    idx = df.columns[i]
    thing = df[idx]
    ret_tot.append(list(thing))
```

```python
N = len(ret_tot[0])
print(sum(ret_tot[:][0]))

returns = [0]*N

for i in range(157):
    returns += ret_tot[i][:]
new_returns = [x/157 for x in returns]
```

```python
plt.plot(new_returns)
```
