```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# from pandas.table.plotting import table



AAPL = pd.read_csv(r'../data/AAPL.csv', index_col = 0)
IBM = pd.read_csv(r'../data/IBM.csv', index_col = 0)
JPM = pd.read_csv(r'../data/JPM.csv', index_col = 0)
DJI = pd.read_csv(r'../data/DJI.csv', index_col = 0)

sec = ["AAPL", "IBM", "JPM", "DJI"]
```

# 1 day return

```python
AAPL['1DReturn'] = AAPL['Adj Close'].pct_change()
IBM['1DReturn'] = IBM['Adj Close'].pct_change()
JPM['1DReturn'] = JPM['Adj Close'].pct_change()
DJI['1DReturn'] = DJI['Adj Close'].pct_change()
# print(AAPL['1DReturn'])
```

```python
stats = pd.DataFrame(columns=sec)
stats
```

```python
pd.set_option('display.float_format', lambda x: '%.6f' % x)


onedret = pd.DataFrame(columns=sec)

mode = "1DReturn"
mean = []

mean.append(np.mean(AAPL[mode]))
mean.append(np.mean(IBM[mode]))
mean.append(np.mean(JPM[mode]))
mean.append(np.mean(DJI[mode]))


median = []

median.append(AAPL[mode].median())
median.append(IBM[mode].median())
median.append(JPM[mode].median())
median.append(DJI[mode].median())


stddev = []
stddev.append(np.std(AAPL[mode]))
stddev.append(np.std(IBM[mode]))
stddev.append(np.std(JPM[mode]))
stddev.append(np.std(DJI[mode]))
# print(stddev)

s = pd.Series(mean, index=stats.columns)
s.name = mode+"_mean"
onedret = onedret.append(s)

s = pd.Series(median, index=stats.columns)
s.name = mode+"_median"
onedret = onedret.append(s)

s = pd.Series(stddev, index=stats.columns)
s.name = mode+"_stddev"
onedret = onedret.append(s)

onedret
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

modes = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
for mode in modes:
    mean = []

    mean.append(np.mean(AAPL[mode]))
    mean.append(np.mean(IBM[mode]))
    mean.append(np.mean(JPM[mode]))
    mean.append(np.mean(DJI[mode]))


    median = []

    median.append(np.median(AAPL[mode]))
    median.append(np.median(IBM[mode]))
    median.append(np.median(JPM[mode]))
    median.append(np.median(DJI[mode]))


    stddev = []
    stddev.append(np.std(AAPL[mode]))
    stddev.append(np.std(IBM[mode]))
    stddev.append(np.std(JPM[mode]))
    stddev.append(np.std(DJI[mode]))
    # print(stddev)

    s = pd.Series(mean, index=stats.columns)
    s.name = mode+"_mean"
    stats = stats.append(s)

    s = pd.Series(median, index=stats.columns)
    s.name = mode+"_median"
    stats = stats.append(s)

    s = pd.Series(stddev, index=stats.columns)
    s.name = mode+"_stddev"
    stats = stats.append(s)
    
#     df.loc[:4].style.background_gradient(cmap='viridis')
# color='#d65f5f'
# stats.loc[:].style.background_gradient(cmap='viridis')
stats
```
